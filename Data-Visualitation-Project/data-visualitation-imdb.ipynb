# Analisis IMDb Score Berdasarkan Negara dan Faktor Lainnya

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load dataset

```python
dataset = pd.read_csv("dataset/dataset.csv")
```

# 1. Perkembangan Rata-rata IMDb Score Berdasarkan Negara
Grafik ini menunjukkan perkembangan rata-rata IMDb score dari film berdasarkan negara yang paling banyak berkontribusi dalam dataset. Data diambil dari lima negara dengan jumlah film terbanyak.

```python
top_countries = dataset['country'].value_counts().index[:5]  
plt.figure(figsize=(10, 5))
for country in top_countries:
    country_data = dataset[dataset['country'] == country].groupby('country')['imdb_score'].mean()
    plt.plot(country_data.index, country_data.values, marker='o', label=country)

plt.xlabel("Negara")
plt.ylabel("Rata-rata IMDb Score")
plt.title("Perkembangan Rata-rata IMDb Score Berdasarkan Negara")
plt.legend()
plt.grid(True)
plt.show()
```

# 2. 10 Negara dengan Rata-rata IMDb Score Tertinggi
Diagram batang ini menunjukkan 10 negara dengan rata-rata IMDb score tertinggi berdasarkan data dalam dataset.

```python
plt.figure(figsize=(10, 5))

country_scores = dataset.groupby('country')['imdb_score'].mean().sort_values(ascending=False)
top_10_countries = country_scores[:10]

sns.barplot(x=top_10_countries.values, y=top_10_countries.index, hue=top_10_countries.index, palette="viridis", legend=False)

plt.xlabel("Rata-rata IMDb Score")
plt.ylabel("Negara")
plt.title("10 Negara dengan Rata-rata IMDb Score Tertinggi")
plt.show()
```

# 3. Distribusi IMDb Score Class
Diagram pie ini menunjukkan distribusi kelas IMDb score berdasarkan pengelompokan tertentu dalam dataset.

```python
plt.figure(figsize=(7, 7))
score_class_counts = dataset['imdb_score_class'].value_counts()
plt.pie(score_class_counts, labels=score_class_counts.index, autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title("Distribusi IMDb Score Class")
plt.show()
```

# 4. Distribusi IMDb Score
Histogram ini menunjukkan distribusi nilai IMDb score di dalam dataset.

```python
plt.figure(figsize=(10, 5))
sns.histplot(dataset['imdb_score'], bins=20, kde=True, color="blue")
plt.xlabel("IMDb Score")
plt.ylabel("Jumlah Film")
plt.title("Distribusi IMDb Score")
plt.show()
```

# 5. Hubungan Budget dan Gross Revenue
Scatter plot ini menggambarkan hubungan antara budget dan gross revenue dari film dalam dataset dengan skala logaritmik.

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(x=dataset['budget'], y=dataset['gross'], alpha=0.5, color="red")
plt.xlabel("Budget (USD)")
plt.ylabel("Gross Revenue (USD)")
plt.title("Hubungan Budget dan Gross Revenue")
plt.xscale('log')  
plt.yscale('log')
plt.grid(True)
plt.show()
```
